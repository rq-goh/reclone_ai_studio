```python
import os, base64

creds = f"{os.environ['GITHUB_USER']}:{os.environ['GITHUB_TOKEN']}".encode("utf-8")
basic_auth = base64.b64encode(creds).decode("utf-8")
os.environ["GITHUB_BASIC_AUTH"] = basic_auth

print("Basic auth header prepared.")
```

```python
%%bash
set -e
cd "/content/drive/MyDrive/reclone_ai_studio/reclone_ai_studio"

git -c http.extraheader="Authorization: Basic ${GITHUB_BASIC_AUTH}" push --force -u origin main
```
